```python
#### volvemos todo a minuscula
data = pd.read_csv('datos_finales.csv', engine='python',sep="|") ## Lectura de los datos
regex = re.compile('[^A-Za-z]')
## dejo solo la letras
def texCC(x,rg): return rg.sub(' ',str(x))
def mas(x): return lower(x)
data_F = data.assign(caption = data.caption.map(lambda p: texCC(p, regex)))
### paso todo a minisculas
prueba = data_F['caption']
prueba = prueba.str.lower()
data_F['caption']=prueba

#### en principio con 1 a 10 frecuecnias
data_F.head(4)
```

* Tomamos los comentarias que tienen entre 1 y 7 palabras, puesto que consideramos que los comentarios con
mas palabrasno ofrecen suficiente informacion para tenerla en cuenta.
* El total de imagenes que nos quedaro despues de realizar el filtro es : 33224

```python
# Cantidad de palabras por comentario
plt.rcParams['figure.figsize'] = [10, 5]
def LENW(x):return len(tknzr.tokenize(str(x)))
data_F = data_F.assign(len_word = data_F.caption.map(LENW))
data_F= data_F[(data_F.len_word>=1) & (data_F.len_word<=7)]
#prueba.head(20)

pd.crosstab(data_F.len_word, columns='count').plot.bar()
plt.title("Cantidad de palabras por comentario")
plt.show()
```

```python
regex2 = re.compile('[^A-Za-z0-9]')
data_F2 = data_F.assign(caption = data_F.caption.map(lambda p: texCC(p, regex2)))
def NGRAM(x,n):
    token=nltk.word_tokenize(str(x))
    bigrams=ngrams(token,n)
    return list(bigrams)
def Table_NGRAM(dat,n):
    n_grams = dat.caption.map(lambda p: NGRAM(p,n))
    n_grams = sum(n_grams,[])
    table_n = pd.DataFrame(FreqDist(n_grams), index =[0]).T
    table_n.columns = ['Count']
    table_n = table_n.sort_values('Count', ascending = False)
    return table_n


table_n1 = Table_NGRAM(data_F2,1)
table_n1 = table_n1.assign(palabra=table_n1.index)
regex2 = re.compile('[^A-Za-z0-9]')
table_n1 = table_n1.assign(palabra= table_n1.palabra.map(lambda p: texCC(p, regex2)))
plt.rcParams['figure.figsize'] = [10, 5]
def cuento(x):return len(str(x.strip()))
table_n1 = table_n1.assign(nletra= table_n1.palabra.map(cuento))
table_n1= table_n1[table_n1.nletra>3] 
table_n1=table_n1.drop(['nletra','palabra'],axis=1)

```
